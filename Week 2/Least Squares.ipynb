## An overdetermined system

Consider the following system of equations:

\begin{align}
2x - y &= 2 \\ x + 2y &= 1 \\ x + y &= 4
\end{align}

Rewriting them to plot x vs. y, we get

\begin{align}
y &= 2x - 2 \\
y &= \frac{-1}{2}x + \frac{1}{2} \\
y &= -x + 4 \\
\end{align}

In linear algebgra, we're often looking for the point in which a system of linear equations all intercept one another. In an **overdetermined** system, however, this is impossible: we have more equations than unknown terms, and consequently become **overconstrained**, and unable find that single intercept.
